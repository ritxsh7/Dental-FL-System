```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random training metrics for two models
epochs = np.arange(1, 51)

# Model 1 accuracy (from 30% to 87%)
model1_accuracy = np.linspace(0.3, 0.87, 50) + np.random.normal(0, 0.05, 50)

# Model 2 accuracy (slightly lower range, from 25% to 75%)
model2_accuracy = np.linspace(0.25, 0.75, 50) + np.random.normal(0, 0.05, 50)

# Plotting the accuracy chart
plt.figure(figsize=(10, 6))
plt.plot(epochs, model1_accuracy, label="WIFA aggregation", color='b', marker='o')
plt.plot(epochs, model2_accuracy, label="FedAvg aggregation ", color='r', marker='o')

# Adding titles and labels
plt.title("Training Accuracy of Two Models Over Aggregation Rounds")
plt.xlabel("Rounds")
plt.ylabel("Accuracy")
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

```python
# Define the data for the bar graph
categories = ['H.E.', 'WIFA (Proposed method)','FedAvg', 'FedProx', 'FedNova' ]
global_model_accuracy = [0.90, 0.88, 0.82, 0.78, 0.76 ]  # Hypothetical global model accuracies
client_own_accuracy = [0.85, 0.83, 0.79, 0.70, 0.72 ]  # Hypothetical client-all accuracies
client_all_accuracy = [ 0.78, 0.80, 0.70,0.68, 0.71]  # Hypothetical client-own accuracies

# Set the x-axis positions for each category
x_pos = np.arange(len(categories))

# Set the width of each bar
bar_width = 0.25

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the bars for each category
ax.bar(x_pos - bar_width, global_model_accuracy, bar_width, label='Global Model Accuracy', color='b')
ax.bar(x_pos, client_all_accuracy, bar_width, label='Client All Accuracy', color='g')
ax.bar(x_pos + bar_width, client_own_accuracy, bar_width, label='Client Own Accuracy', color='r')

# Add labels and title
ax.set_xlabel('Models')
ax.set_ylabel('Accuracy')
ax.set_title('Comparison of Model Accuracies (Global, Client All, and Client Own)')

# Set the x-axis labels
ax.set_xticks(x_pos)
ax.set_xticklabels(categories)

# Add a legend
ax.legend()

# Adjust spacing to place FedWt to the right of the other models
ax.set_xlim(-0.5, len(categories) - 1 + 1.5)

# Show the plot
plt.tight_layout()
plt.show()
```
